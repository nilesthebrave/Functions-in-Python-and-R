## Problem 1

ðŸŽ¯Write a function that takes in an arbitrary list of numeric triples (tuples of size 3), and returns a new list containing the maximum elements from each

```python
# Define function to return maximum elements from a list of tuples
def MaxElement(tupleList):
    maxList = [] # declare an empty list
    for i in tupleList: # loop through the list of tuples
        for j in i: # loop through individual elements of the tuples
            if j == max(i): # check if the current element is the highest element of the tuple
                maxList.append(j) # if yes, append it to the list
    return maxList
```

```python
tup = [(3,5,4)] # arbitrary list of tuples
MaxElement(tup)
```

## Problem 2

ðŸŽ¯Write a function that takes no arguments, gets the current time, and returns to the caller the integer number of minutes left in the current day (til midnight ðŸŒƒ)

```python
# importing datetime because I'm not sure I know where to begin otherwise
from datetime import datetime

# define a function to convert time to minutes and return time left in day
def TimeLeft():
    dt = datetime.now().time() # get current time
    dt = dt.strftime('%H/%M') # convert that time into a string, including only hours and minutes
    
    hours,minutes = dt.split('/') # split that string into two separate strings
    hours,minutes = int(hours),int(minutes) # convert both strings to ints
    hours = hours*60 # convert str hours into minutes
    
    currentMinutes = hours+minutes # add hours and minutes together
    min2Midnight=1440-currentMinutes # 1440 minutes in a day, subtract current minutes
    
    return min2Midnight # time left
```

```python
TimeLeft()
```
